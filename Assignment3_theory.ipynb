###1. What is the difference between a function and a method in Python?


In Python, both functions and methods are blocks of reusable code that perform a specific task, but they differ primarily in how they are associated with objects.

**Function:-** Classes are not required to define a function.

Functions are independent of objects as they are not defined within a class.

Functions can be invoked just by calling their names.

Functions perform their operation on the data that the user passes as arguments or parameters.

These are independent entities of a Python code.

‘Self’ parameter is not required in the case of functions. Functions may have zero or many arguments.

Functions can be used for both object-oriented and non-object-oriented programming languages. Non-object-oriented programming languages are referred to as procedural programming languages.

**Method:-** Methods are always defined within a class.

Methods are linked to the objects of the class in which they are defined.

A method cannot be invoked by calling its name alone. The class to which it belongs should be invoked.

Methods can execute the logic with the data enclosed in the object of the class to which it is associated.

These are the class dependent entities of a Python code.

The first argument of a method should be ‘self’.

Methods are applicable only for object-oriented programming languages.

```python
# Define a function
def factorial(n):
    return 1 if (n==1 or n==0) else n * factorial(n-1);
n = int(6)
print("The factorial of",n,"is",factorial(n))
```

```python
# Define a method
class Course(object):
   def my_method(self):
      print("This is method")
name = Course()
name.my_method()
```

###2. Explain the concept of function arguments and parameters in Python?

In Python, the terms arguments and parameters are often used interchangeably but they have distinct meanings when it comes to functions.

**Arguments:-** Arguments are the actual values or data passed to a function when it is called. These values are assigned to the corresponding parameters in the function definition.

Arguments provide the input values that the function operates on.

***Parameters:-*** Parameters are the variable names listed inside the parentheses in the function definition. They act as placeholders for the values that the function will receive when it is called.

Parameters define the input that the function expects.

**Types of Parameters (and Corresponding Arguments)**

**Positional Parameters:-** Parameters that are assigned values based on their position in the function call.

**Default Parameters:-** Parameters that have a default value, which is used if no argument is provided for that parameter.

**Keyword (Named) Parameters:-** Parameters that are explicitly assigned a value by name in the function call.

**Arbitrary Positional Parameters (*args):-** Allows you to pass a variable number of positional arguments. These arguments are captured as a tuple.

```python
# Argument
def sum(a,b):
  print(a+b)
sum(5,6) # Here the values 5,6 are arguments
```

```python
# Here a,b are the parameters
def sum(a,b):
  print(a+b)
sum(5,8)
```

```python
def add(x, y):  # 'x' and 'y' are positional parameters
    return x + y
add(7, 9)  # Arguments 7 and 9 are assigned to x and y, respectively
```

###3. What are the different ways to define and call a function in Python?

Functions can be defined in multiple ways, offering flexibility in how you organize and structure your code.

**Regular Function Definition:-** This is the most common way to define a function using the def keyword.

```python
def greet(name):
    return f"Hello {name}"
greet("Python")
```

**Function with Default Arguments:-** You can define a function with default parameters, where default values are used if no arguments are provided.

```python
def greet(name="Guest"):
    return f"Hello {name}"
greet()
greet("Python")
```

**Function with Arbitrary Positional Arguments (*args):-** You can define a function that accepts a variable number of positional arguments using *args.

```python
def add_all(*a):
    return sum(*a)
add_all(4, 2)
add_all(8, 10)
```

**Function with Arbitrary Keyword Arguments (**kwargs):-** A function can also accept a variable number of keyword arguments using **kwargs.

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_info(name="Rony", age=26, city="Australia")
```

**Lambda Functions (Anonymous Functions):-** Lambda functions are small, anonymous functions that can be defined using the lambda keyword. They are typically used for short, simple functions.

```python
square = lambda x: x ** 2
square(5)
```

**Higher-Order Functions:-** A function that takes another function as an argument or returns a function is called a higher-order function. These functions are commonly used in Python with functions like map(), filter(), etc.

```python
def apply_func(func, value):
    return func(value)
apply_func(lambda x: x ** 2, 4)
```

**Recursion:-** A recursive function is a function that calls itself, typically used for solving problems that can be broken down into smaller, similar subproblems.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
factorial(4)
```

###4. What is the purpose of the `return` statement in a Python function?

A return statement is used to end the execution of the function call and “returns” the result (value of the expression following the return keyword) to the caller.

The statements after the return statements are not executed. If the return statement is without any expression, then the special value None is returned.

A return statement is overall used to invoke a function so that the passed statements can be executed.

The return statement in a Python function serves the following key purposes:

**Returning a Value:-** The primary purpose of the return statement is to send a value back to the caller of the function. This allows the function to output a result that can be used elsewhere in the code.

```python
def add(a, b):
    return a + b
result = add(5, 7)
print(result)
```

**Ending a Function:-** When a return statement is executed, it immediately terminates the function, and any code after the return statement in the function body will not be executed.

**Returning Multiple Values:-** you can return multiple values from a function by separating them with commas. These values are returned as a tuple.

```python
def get_coordinates():
    return 20, 30
x, y = get_coordinates()
print(x, y)
```

**Returning Nothing (None):-** If a return statement is used without an expression, or if a function doesn't include a return statement, it implicitly returns None. This is the default behavior for functions that don't explicitly return a value.

```python
def no_return():
    print("This function has no return value")
result = no_return()
print(result)
```

###5. What are iterators in Python and how do they differ from iterables?

iterators and iterables are closely related concepts, but they serve different purposes. Understanding the distinction between them is important when working with loops, generators, and collections in Python.

**Iterators:-** An iterator is an object that represents a stream of data. It allows you to traverse through the elements of an iterable, one at a time, using the __next__() method (or next() in Python). An iterator remembers where it is during the iteration process. Used to fetch elements one by one using next() or a for loop. The object returned by iter() when called on an iterable.

An iterator is an object that implements two methods:-

__iter__() to return itself (so that it can be used in a loop).

__next__() to return the next value in the sequence and raise StopIteration when there are no more elements.

Once an iterator is exhausted (i.e., all elements are iterated), it cannot be reused unless you explicitly create a new iterator.

```python
# Iterable (a list)
my_list = [10, 20, 30, 40]
iterator = iter(my_list)

print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

**Iterables:-** An iterable is any Python object that can return an iterator, which allows you to loop over (iterate through) its elements one by one. Common examples of iterables include lists, tuples, strings, dictionaries, sets, and even custom objects that implement the __iter__() method. An object is considered an iterable if it can return an iterator via the __iter__() method.
Used with a for loop to iterate over elements. Does not track its position in iteration.

Examples of Iterables:

Lists: my_list = [1, 2, 3]

Strings: my_string = "hello"

Tuples: my_tuple = (1, 2, 3)

```python
my_list = [1, 2, 3, 4, 5]
for item in my_list:
    print(item)
```

###6. Explain the concept of generators in Python and how they are defined?

Generators are a special type of iterator that allow you to lazily evaluate a sequence of values. Instead of returning all the values at once (like a list), generators produce items one at a time and only when requested. This makes them memory-efficient, especially when working with large datasets or streams of data.

A generator is defined using a function that contains one or more yield statements. When the generator function is called, it doesn't run the code immediately. Instead, it returns a generator object, which can be iterated over.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
counter = count_up_to(2)
print(next(counter))
print(next(counter))
print(next(counter)) # Raises StopIteration since the sequence is exhausted
```

***How Generators Work:***
Each time yield is executed, the generator pauses its state (including local variables), and the value is sent back to the caller.

The next time the generator is called (using next()), it resumes from where it left off, continuing the execution until the next yield.

When there are no more values to yield, the generator raises a StopIteration exception to signal the end of iteration, just like an iterator.

Generators are often used within loops, as they can yield an indefinite number of values lazily.

```python
# Generator expression to generate squares of numbers
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib_gen = fibonacci()

for _ in range(3):
    print(next(fib_gen))

gen_exp = (x ** 2 for x in range(3))
print(next(gen_exp))
print(next(gen_exp))
print(next(gen_exp))
```

###7. What are the advantages of using generators over regular functions?



Generators offer several advantages over regular functions in Python, particularly when it comes to performance, memory efficiency, and ease of use in certain scenarios.

**Key advantages of using generators over regular functions:**

**Memory Efficiency:** Generators do not store all values in memory, which is ideal for large or infinite sequences.

**Lazy Evaluation:** Values are generated only when needed, improving performance.
Handling Infinite Sequences: Generators can produce infinite sequences without consuming memory.

**Improved Performance:** Generators can return values faster by yielding results incrementally.

**Simplified Iterators:** Using yield simplifies the creation of iterators compared to manually implementing them.

**Streaming and Pipelines:** Generators are excellent for pipelining and processing data in chunks.

**State Persistence:** Generators retain their state, allowing for efficient control of flow during iteration.

```python
def fibonacci():
    a, b = 1, 2
    while True:
        yield a
        a, b = b, a + b
gen = fibonacci()
print(next(gen))
print(next(gen))
```

###8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function defined using the lambda keyword. It is a concise way to create simple functions without needing to formally define them using the def keyword. Lambda functions are typically used for short, throwaway operations where defining a full function might seem unnecessary or cumbersome.

Lambda functions are typically used in situations where a simple function is required for a short period, and defining a regular function using def would be overkill. Common use cases include:

As an argument to higher-order functions (functions that take other functions as arguments), such as map(), filter(), and sorted().

**In sorted() or sort() for custom sorting:** You can use lambda functions to specify custom sorting logic. The sorted() function and the sort() method accept a key parameter, which allows sorting based on a computed key.

**In reduce():** The reduce() function from the functools module applies a binary function (like summing or multiplying) cumulatively to the elements of an iterable. Lambda functions are often used for this.

**Anonymous functions in GUI applications or event handling:** Lambda functions can be used as callbacks or small functions in graphical user interfaces (GUIs) or when responding to events.

**Concise syntax:** Lambda functions allow you to define small, single-use functions in a more compact and readable way.

**Anonymous:** Lambda functions don't require a name, making them useful when you don't need to reuse the function elsewhere.

**Function as a first-class object:** Python treats functions as first-class objects, meaning you can pass them as arguments to other functions, return them from functions, or store them in variables. Lambda functions make this process simpler for short functions.

```python
add = lambda x, y: x + y
print(add(5, 6))
```

```python
# Example with map()
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

###9. Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is used to apply a given function to every item in an iterable (like a list, tuple, etc.) and returns an iterator (or map object) containing the results. It allows for efficient and concise transformation of data by applying a function to each element in a sequence without the need for an explicit loop.

Syntax of map(): map(function, iterable, ...)

function: A function that takes one or more arguments and is applied to each element in the iterable(s).

iterable: One or more iterables (like lists, tuples, or sets) whose elements will be processed by the function.

The map() function applies the function to each item of the iterable, yielding results one at a time (lazy evaluation). If multiple iterables are passed, the function must accept the same number of arguments as there are iterables, and map() will apply the function to corresponding elements from all iterables.

```python
# Converting the map object to a list to view the results
def square(x):
    return x ** 2
numbers = [1, 2, 3,]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

***Usage of map():***

The map() function is particularly useful when you want to:

Transform elements of an iterable: Apply a transformation to each element (like squaring numbers, converting strings to uppercase, etc.).

Apply the same function to multiple sequences: Perform element-wise operations on two or more iterables, like adding numbers from two lists.

Work with lazy evaluation: map() returns an iterator, which means the function is applied lazily, yielding results only as they are needed.

###10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

map(), reduce(), and filter() are higher-order functions that allow you to apply functions to sequences (like lists or tuples). Although they are all related to functional programming, each of these functions serves a different purpose.

*-> map()**

The map() function applies a given function to each item in an iterable (or multiple iterables) and returns an iterator (or map object) of the results.

**Purpose:**
To transform or modify each element in an iterable using a function.

**Syntax:**

    map(function, iterable, ...)

function: A function that is applied to each element of the iterable(s).

iterable: One or more iterables (such as lists, tuples, etc.).

**Key Characteristics:**

Transforms each element individually.

Can be used with multiple iterables, applying the function to corresponding elements.

Returns an iterator (map object), which can be converted to a list, tuple, etc.

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))
```

***filter()***

The filter() function filters elements from an iterable based on a condition provided by a function. The function should return True or False, and filter() keeps only the elements for which the function returns True.

**Purpose:**
To select or filter elements from an iterable based on a condition.

**Syntax:**

    filter(function, iterable)

function: A function that returns True or False for each element.

iterable: The iterable to be filtered.

**Key Characteristics:**

Transforms each element individually.

Can be used with multiple iterables, applying the function to corresponding elements.

Returns an iterator (map object), which can be converted to a list, tuple, etc.



```python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

***->reduce()***

The reduce() function applies a function of two arguments cumulatively to the items of an iterable, reducing the iterable to a single value. It’s commonly used for aggregating results, such as summing or multiplying all elements in a list.

**Purpose:**
To reduce a sequence of elements to a single result by applying a binary function cumulatively.

**Syntax:**

    from functools import reduce
    reduce(function, iterable, initializer)

function: A function that takes two arguments and returns a single value.

iterable: The iterable to reduce.

initializer (optional): A starting value for the reduction (default is the first element of the iterable).

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```
