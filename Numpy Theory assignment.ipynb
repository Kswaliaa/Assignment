# Ans 7

**vstack()** stacks vertical, increase in number of rows.

**hstack()** stacks horizontal, increase in number of columns.


**Examples Below of vstack() and hstack()**

```python
import numpy as np

array1 = np.array([[1, 2, 3],[6, 8, 9]])
array2 = np.array([[4, 5, 6],[2, 5, 10]])

# Vertical
result_vertical = np.vstack((array1, array2))

print("result_vertical:")
print(result_vertical)
```

```python
import numpy as np

array1 = np.array([[1, 2, 3],[6, 8, 9]])
array2 = np.array([[4, 5, 6],[2, 5, 10]])

# Horizontal
result_horizontal = np.hstack((array1, array2))

print("result_horizontal:")
print(result_horizontal)
```

## **flipud()**

**Purpose:** Flips array are vertically.

**Applicable Dimension:** affects 2D arrays but it's been applied to higher-dimensional arrays, where it flip with first axis.

**Behavior:** In each column of the array, the elements are reverse in rows.



## Effects on Various Array Dimensions

**1**)**2D Arrays:**

fliplr() reverses the  elements in each row of order.

flipud() reverses the order of the rows.

2)1D Arrays:**

fliplr() behaves as same as flipud() for 1D arrays by flipping them to reverse order of elements.

**3)Higher-Dimensional Arrays:**

Vectorization and broadcasting are two key concepts in NumPy that greatly affect the efficiency of array operations.

**Vectorization:-** Vectorization is to process the convert operations that  typically be in a loop into single operation that  perform on entire arrays once.This is because of NumPy operations are added into C code which runs much fast than Python loops.

**Benefits:**

**Speed:** By avoiding Python for loops and leveraging low-level implementations, vectorized operations are faster.

**Code Simplicity:** Vectorization leads to cleaner are more readable code operations can be expressed in fewer lines.

** Broadcasting**:- It is strong mechanism which allows NumPy to work with arrays has  different shapes during arithmetic operations it enables automatic expansion of smaller arrays to match with the shape of larger arrays without copying data, it saves memory.

**Working Process**

**Dimension Alignment:** NumPy compare the shapes of the two arrays, if they are different, it tries to by "stretching" the smaller array along the dimensions that are missing.

## Contribution to Efficient Array Operations

**Memory Efficiency:** Broadcasting allows for operations on arrays of different shapes without creating largemarrays, it is particularly beneficial working with large datasets.

**Simplicity and Clarity:** Code becomes easy to understand. Vectorized operations and broadcasting it replace loop it construct with simple arithmetic expressions for making the code clearer.

**Reduced Computation Time:**  vectorization and broadcasting both minimize the need for explicit loops in Python, which is slow due to interpreter overhead.
